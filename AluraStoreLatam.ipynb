### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda2.head()
tienda3.head()
tienda4.head()
```

#1. Análisis de facturación



```python
ingreso_total1 = tienda['Precio'].sum()
print("Ingreso total:", ingreso_total1)
ingreso_total2 = tienda2['Precio'].sum()
print("Ingreso total:", ingreso_total2)
ingreso_total3 = tienda3['Precio'].sum()
print("Ingreso total:", ingreso_total3)
ingreso_total4 = tienda4['Precio'].sum()
print("Ingreso total:", ingreso_total4)
```

# 2. Ventas por categoría

```python
ventas_categoria = tienda.groupby(['Categoría del Producto']).size().sort_values(ascending=False)
ventas_categoria
```

```python
ventas_categoria2 = tienda2.groupby(['Categoría del Producto']).size().sort_values(ascending=False)
ventas_categoria2
```

```python
ventas_categoria3 = tienda3.groupby(['Categoría del Producto']).size().sort_values(ascending=False)
ventas_categoria3
```

```python
ventas_categoria4 = tienda4.groupby(['Categoría del Producto']).size().sort_values(ascending=False)
ventas_categoria4
```

```python

df_ventas = pd.concat(
    [
        ventas_categoria.rename("Tienda 1"),
        ventas_categoria2.rename("Tienda 2"),
        ventas_categoria3.rename("Tienda 3"),
        ventas_categoria4.rename("Tienda 4")
    ],
    axis=1
)
df_ventas
```

```python
def resumen_tienda(df, nombre_tienda):
    cantidad = df.groupby("Categoría del Producto").size()
    valor = df.groupby("Categoría del Producto")["Precio"].sum()
    return pd.concat([cantidad.rename(f"{nombre_tienda}_Cantidad"),
                      valor.rename(f"{nombre_tienda}_Valor")], axis=1)

t1 = resumen_tienda(tienda, "Tienda1")
t2 = resumen_tienda(tienda2, "Tienda2")
t3 = resumen_tienda(tienda3, "Tienda3")
t4 = resumen_tienda(tienda4, "Tienda4")

df_comparacion = pd.concat([t1, t2, t3, t4], axis=1)
df_comparacion
```
